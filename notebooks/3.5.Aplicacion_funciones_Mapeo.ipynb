```python
import pandas as pd
```

```python
data = {
    'Producto': ['Manzanas', 'Naranjas', 'Plátanos', 'Uvas', 'Peras'],
    'Precio': [100, 80, 60, 120, 90]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
def descuento(x):
    return x * 0.85
```

```python
df['Precio_Descuento'] = df['Precio'].apply(descuento)
```

```python
df
```

```python
df['Producto'] = df['Producto'].apply(lambda x: x.upper())
```

```python
df
```

```python
df['Producto'] = df['Producto'].map({'Manzanas': 'Apples', 'Naranjas': 'Oranges'})
df
```

```python
df = pd.DataFrame(data)
df
```

```python
df = df.applymap(lambda x: x.lower() if type(x) == str else x)
df
```

```python
df['Descuento'] = [10, 15, 10, 20, 5]
```

```python
df
```

```python
df['Precio_Final'] = df.apply(lambda row: row['Precio'] * (100 - row['Descuento']) / 100, axis=1)
df
```
