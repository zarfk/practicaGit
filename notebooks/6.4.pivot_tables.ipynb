```python
import pandas as pd
```

```python
data = {
    'Empleado': ['Juan', 'Ana', 'Pedro', 'Luis', 'Maria', 'Ana', 'Pedro', 'Luis', 'Maria'],
    'Departamento': ['Ventas', 'Marketing', 'Ventas', 'IT', 'Marketing', 'IT', 'Ventas', 'Marketing', 'Ventas'],
    'Salario': [50000, 60000, 55000, 58000, 62000, 60000, 55000, 58000, 62000],
    'A単os_Experiencia': [5, 7, 6, 8, 9, 7, 6, 8, 9]
}

df = pd.DataFrame(data)
display(df)
```

```python
pivot_table = df.pivot_table(values='Salario', index='Departamento', aggfunc=['sum','mean','min','max'])
display(pivot_table)
```

```python
pivot_table_multi = df.pivot_table(values='Salario', index=['Departamento', 'Empleado'], columns='A単os_Experiencia', aggfunc='sum')
display(pivot_table_multi)
```

```python
cross_tab = pd.crosstab(df['Departamento'], df['Empleado'])
display(cross_tab)
```

```python
cross_tab_agg = pd.crosstab(df['Departamento'], df['Empleado'], values=df['Salario'], aggfunc='sum', margins=True)
display(cross_tab_agg)
```

```python
cross_tab_norm = pd.crosstab(df['Departamento'], df['Empleado'], normalize='index')
cross_tab_norm
```
