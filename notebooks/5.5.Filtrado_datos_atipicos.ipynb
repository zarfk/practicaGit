```python
import pandas as pd
```

```python
data = {
    'Empleado': ['Juan', 'Ana', 'Pedro', 'Luis', 'Maria','Pedro'],
    'Salario': [50000, 60000, 55000, 58000, 120000,500],  # María tiene un salario atípico
    'Años_Experiencia': [5, 7, 6, 8, 9,1]
}

df = pd.DataFrame(data)
df
```

```python
media_salario = df['Salario'].mean()
std_salario = df['Salario'].std()
umbral = 1

df['Es_Atipico'] = abs(df['Salario'] - media_salario) > umbral * std_salario
df
```

```python
Q1 = df['Salario'].quantile(0.25)
Q3 = df['Salario'].quantile(0.75)
IQR = Q3 - Q1

umbral_inferior = Q1 - 1.5 * IQR
umbral_superior = Q3 + 1.5 * IQR

df['Es_Atipico'] = (df['Salario'] < umbral_inferior) | (df['Salario'] > umbral_superior)
df
```

```python
df_sin_atipicos = df[~df['Es_Atipico']]
df_sin_atipicos
```

```python
df.loc[df['Es_Atipico'], 'Salario'] = media_salario
df
```
